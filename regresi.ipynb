```python
# Import Library yang diperlukan Scipy dan Pandas
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import sklearn
```

```python
# Import dataset
dataset = pd.read_csv('datanilai.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
print(x)
```

```python
print(y)
```

```python
dataset.keys()
```

```python
dataset.shape
```

```python
# Menampilkan isi sebagian dataset
data = pd.DataFrame(dataset)
data.head()
```

```python
# Split Dataset menjadi Training set dan Testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=0)
```

```python
# Melakukan Fitting Simple Linier Regression pada Training Set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() 
regressor.fit(x_train, y_train)
```

```python
# Memprediksi hasil Test-Set
y_pred = regressor.predict(x_test)
```

```python
#visualisasi data
plt.scatter(data.SAT, data.GPA)
plt.xlabel("SAT")
plt.ylabel("GPA")
plt.title("Grafik SAT vs GPA")
plt.show
```

```python
#visualisai hasil prediksi pada training set
#ukuran plot
plt.figure(figsize=(10,8))

#Biru adalah data observasi 
plt.scatter(x_train, y_train, color = 'blue')

#Garis merah adalah hasil prediksi dari machine learning
plt.plot(x_train, regressor.predict(x_train), color = 'red')

#Memberi Judul dan label
plt.title("SAT terhadap GPA (Training Set)")
plt.xlabel("SAT")
plt.ylabel("GPA")

#menampilkan plot
plt.show()
```

```python
#visualisai hasil prediksi pada test set
#ukuran plot
plt.figure(figsize=(10,8))

#Biru adalah data observasi 
plt.scatter(x_test, y_test, color = 'blue')

#Garis merah adalah hasil prediksi dari machine learning
plt.plot(x_test, regressor.predict(x_test), color = 'red')

#Memberi Judul dan label
plt.title("SAT terhadap GPA (Test Set)")
plt.xlabel("SAT")
plt.ylabel("GPA")

#menampilkan plot
plt.show()
```
